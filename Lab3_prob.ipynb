```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=int(input())
b=int(input())
x=np.arange(a,b+1)
y=[]
for i in range(a,b+1):
  y.append(1/(b-a+1))

plt.bar(x,y,color='maroon',width=0.1)
plt.show()
```

```python
#2
succ=float(input())
x=np.arange(0,2)
y=[1-succ,succ]
plt.title('Bernoulli')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.bar(x,y,width=0.1,color='pink')
```

```python
#3
def fa(c):
  if c==0:
    return 1
  p=1
  for i in range(1,c+1):
    p*=i
  return p

a=int(input())
s=float(input())
x=np.arange(a+1)
y=[]
for i in range(a+1):
  q=fa(a)/(fa(i)*fa(a-i))*(s**i)*((1-s)**(a-i))
  y.append(q)
plt.title('Binomial')
plt.xlabel('X')
plt.ylabel('Probability')
plt.bar(x,y)

```

```python
#4
import math
def fa(c):
  if c==0:
    return 1
  p=1
  for i in range(1,c+1):
    p*=i
  return p

a=int(input())
lam=float(input())
e=2.71828
i=np.arange(a+1)
y=[]
for x in range(a+1):
  q=((e**(-1*lam))*(lam**x))/fa(x)
  y.append(q)
plt.xlabel("no. of trials")
plt.ylabel('Probability')
plt.title('Poisson')
plt.bar(i,y,width=0.50)
plt.show()

```

```python
#5
a=int(input())
p=float(input('succ'))
x=np.arange(a+1)
y=[]
for i in range(a+1):
  q=p*((1-p)**(i-1))
  y.append(q)

plt.xlabel('no.of trials')
plt.ylabel('probability')
plt.title('Geometric PMF')

plt.bar(x,y,width=0.2,color='green')
plt.show()
```
