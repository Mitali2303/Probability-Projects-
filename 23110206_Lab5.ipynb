```python

#importing cv2, numpy and matplotlib.pyplot
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Showing the img 1 also the source image
img_color1 = cv2.imread('/content/grey_1.png',1)
plt.title("Image1")
plt.imshow(img_color1)
plt.axis("off")
plt.show()

# Creating histogram of source image
histogram1=np.zeros(256,dtype=int)
for i in range(img_color1.shape[0]):
  for j in range(img_color1.shape[1]):
    intensity=img_color1[i,j]
    histogram1[intensity]+=1

# Displaying the histogram of source image also called as image1
plt.figure()
plt.bar(range(256),histogram1)
plt.xlabel("Intensity")
plt.ylabel("Frequency of intensities")
plt.title("Histogram1")

# creating another histogram similar to this one and save it which will be used later
hist1=histogram1.copy()

# Calculating cdf of histogram
c1=np.cumsum(histogram1)
histogram1=histogram1/np.sum(histogram1)
cdf1=np.cumsum(histogram1)

# Displaying the graph of cdf of histogram1
plt.figure()
plt.bar(range(256),cdf1)
plt.xlabel("Intensity")
plt.ylabel("Frequency of intensities")
plt.title("CDF of Histogram1")

# Equalising the historgram1
his1=np.zeros(256,dtype=int)
eq=np.zeros(256,dtype=int)
for i in range(256):
  eq[i]=int(np.round(cdf1[i]*255))
  index=int(np.round(cdf1[i]*255))
  his1[index]+=hist1[i]

# Displaying the graph of equalised histogram1
plt.figure()
plt.bar(range(256),his1)
plt.xlabel("Intensity")
plt.ylabel("Frequency of intensities")
plt.title("Equalised Histogram1")

# Creating image of equalised histogram1
for i in range(img_color1.shape[0]):
  for j in range(img_color1.shape[1]):
    img_color1[i,j]=eq[img_color1[i,j]]

# Displaying the created image of equalised histogram1
plt.figure()
plt.imshow(img_color1)
plt.axis('off')
plt.title("Equalised Image1")
########################################

# Showing the img 2 also the target image
img_color2 = cv2.imread('/content/grey_2.png',1)
img_color2 = cv2.resize(img_color2, (img_color1.shape[1], img_color1.shape[0]))
plt.figure()
plt.imshow(img_color2)
plt.title("Image2")
plt.imshow(img_color2)
plt.axis("off")
plt.show()

# Creating histogram of target image
histogram2=np.zeros(256,dtype=int)
for i in range(img_color2.shape[0]):
  for j in range(img_color2.shape[1]):
    intensity=img_color2[i,j]
    histogram2[intensity]+=1

# Displaying the graph of histogram2
plt.figure()
plt.bar(range(256),histogram2)
plt.xlabel("Intensity")
plt.ylabel("Frequency of intensities")
plt.title("Histogram2")

# creating another histogram similar to this one and save it which will be used later
hist2=histogram2.copy()

# Calculating cdf of histogram2
c2=np.cumsum(histogram2)
histogram2=histogram2/(np.sum(histogram2))
cdf2=np.cumsum(histogram2)
histogram2=np.cumsum(histogram2)

# Displaying graph of cdf of histogram2
plt.figure()
plt.bar(range(256),cdf1)
plt.xlabel("Intensity")
plt.ylabel("Frequency of intensities")
plt.title("CDF of Histogram2")

# Equalising Histogram2
his2=np.zeros(256,dtype=int)
eq2=np.zeros(256,dtype=int)
for i in range(256):
  eq2[i]=int(np.round(cdf1[i]*255))
  index=int(np.round(cdf1[i]*255))
  his2[index]+=hist2[i]


# Creating image of equalised histogram2
for i in range(img_color2.shape[0]):
  for j in range(img_color2.shape[1]):
    img_color2[i,j]=eq2[img_color2[i,j]]

# Displaying the created image of equalised histogram1
plt.figure()
plt.imshow(img_color1)
plt.axis('off')
plt.title("Equalised Image2")

# Mapping of source image to target image
map=np.zeros(256,dtype=int)

for i in range(256):
  k=float('inf')
  intensity=0
  for j in range(256):
    diff=np.abs(c1[i]-c2[j])
    if diff<k:
      k=diff
      intensity=j
  map[i]=intensity

# Displaying graph of matched
plt.figure()
plt.bar(range(256),map)
plt.xlabel("Intensity")
plt.ylabel("Frequency of intensities")
plt.title("Mapping of Image1 to Image2")

# Creating the final image of Source image mapped with target image
outimg=np.zeros_like(img_color1)
for i in range(img_color1.shape[0]):
  for j in range(img_color1.shape[1]):
    outimg[i,j]=map[img_color1[i,j]]

# Displaying the mapped final image
plt.figure()
plt.imshow(outimg,cmap='gray')
plt.title("Image1 mapped with Image2")
plt.axis('off')
plt.show()
```
