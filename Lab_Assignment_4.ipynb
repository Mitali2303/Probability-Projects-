```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#1 a)Uniform
a=int(input())
b=int(input())
c=int(input())
d=int(input())
N=1000
X=np.linspace(a-10,b+10,N)
T=np.linspace(c-10,d+10,N)
def uniform(x,a,b):
  if x<a or x>b:
    return(0)
  if b==a:
    return(1)
  return(1/(b-a))
Y=np.array([uniform(x,a,b) for x in X])
y=np.array([uniform(t,c,d) for t in T])
plt.plot(X,Y,label=f'graph 1')
plt.plot(T,y,label=f'graph 2')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.legend(loc='upper right')
plt.fill_between(X, Y, alpha=0.1, color='g')
plt.fill_between(T,y,alpha=0.2,color='pink')
plt.title('pdf of continuous uniform function')
plt.show()
```

```python
#Q2 b1) cdf gaussian
N=1000
X=np.linspace(-10,10,1000)

s=0
a=int(input())
b=int(input())
nu=(a+b)/2
t=np.linspace(a,b,N)
for i in t:
  s+=(i-nu)**2
sd=math.sqrt(s/len(t))
def Gauss(x,sd,nu):
  return (1/2)*(1+math.erf((x-nu)/math.sqrt(2)*sd))

Y=np.array([Gauss(x,sd,nu) for x in X])
plt.plot(X,Y)
plt.ylabel("P(X<=x)")
plt.xlabel("x")
plt.title("Cdf for gaussian distribution")
plt.show()
```

```python
def fa(c):
  if c==0:
    return 1
  p=1
  for i in range(1,c+1):
    p*=i
  return p

a=int(input())
s=float(input())
x=np.arange(a+1)
y=[]
for i in range(a+1):
  q=fa(a)/(fa(i)*fa(a-i))*(s**i)*((1-s)**(a-i))
  y.append(q)
y=np.cumsum(y)
plt.title('Binomial')
plt.xlabel('X')
plt.ylabel('Probability')
plt.bar(x,y)
```

```python
import math
def fa(c):
  if c==0:
    return 1
  p=1
  for i in range(1,c+1):
    p*=i
  return p

a=int(input())
lam=float(input())
e=2.71828
i=np.arange(a+1)
y=[]
for x in range(a+1):
  q=((e**(-1*lam))*(lam**x))/fa(x)
  y.append(q)
y=np.cumsum(y)
plt.xlabel("no. of trials")
plt.ylabel('Probability')
plt.title('Poisson')
plt.bar(i,y,width=0.50)
plt.show()

```
